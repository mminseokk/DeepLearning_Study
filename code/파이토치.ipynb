```python
class BasicModel(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.layer1 = nn.Linear(28,10)

    def forward(self, x):
        out = self.layer1(x)
        return out

model = BasicModel()
print(model)
for p in model.parameters():
    print(p)
```

```python
class BasicModel2(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.layer1 = nn.Linear(784, 256)
        self.layer2 = nn.Linear(256, 64)
        self.layer3 = nn.Linear(64, 10)

    def forward(self, x):
        x = x.view(x.size(0), -1)
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        return out
```

```python
from torch.utils.data import DataLoader
from torch.optim import Adam

model = BasicModel2()
loss_fn = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr=0.1)

train_loader = DataLoader(
    train_data,
    batch_size=32,
    shuffle=True
)
```

```python
correct = 0
for i, data in enumerate(train_loader):
    # if i > 0: break
    images, labels = data
    optimizer.zero_grad()
    output = model(images)
    loss = loss_fn(output, labels)
    loss.backward()
    optimizer.step()
    preds = torch.argmax(output, dim=1)
    corr = (preds == labels).sum()
    acc = corr / len(labels)
    correct += corr
    print(f'loss {loss} acc {acc}')
total_acc = correct / len(train_data)
print(f'total acc {total_acc}')
```

```python
class BasicModel3(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.layer1 = nn.Linear(784, 256)
        self.layer2 = nn.Linear(256, 64)
        self.layer3 = nn.Linear(64, 10)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        x = x.view(x.size(0), -1)
        out = self.layer1(x)
        out = self.relu(out)
        out = self.layer2(out)
        out = self.relu(out)
        out = self.layer3(out)
        return out
```

```python
from torch.optim import Adam
model = BasicModel3()
loss_fn = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr=0.1)
```

```python
def train(model, train_loader, optimizer, loss_fn):
    correct = 0
    losses = []
    for i, data in enumerate(train_loader):
        # if i > 0: break
        images, labels = data
        optimizer.zero_grad()
        output = model(images)
        loss = loss_fn(output, labels)
        loss.backward()
        optimizer.step()
        preds = torch.argmax(output, dim=1)
        corr = (preds == labels).sum()
        acc = corr / len(labels)
        correct += corr
        losses.append(loss.detach().numpy())
        # if i % 10 == 0:
        #   print(f'loss {loss} acc {acc}')
    total_acc = correct / len(train_data)
    print(f'total acc {total_acc}')
    return losses
```

```python
train(model, train_loader, optimizer, loss_fn)
```

```python
class BasicModel5(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.layer1 = nn.Linear(784, 256)
        self.bn1 = nn.BatchNorm1d(256)
        self.layer2 = nn.Linear(256, 64)
        self.bn2 = nn.BatchNorm1d(64)
        self.layer3 = nn.Linear(64, 10)
        self.relu = nn.ReLU(inplace=True)
        self.drop = nn.Dropout(0.3)

    def forward(self, x):
        x = x.view(x.size(0), -1)
        out = self.layer1(x)
        out = self.bn1(out)
        out = self.drop(out)
        out = self.relu(out)
        out = self.layer2(out)
        out = self.bn2(out)
        out = self.drop(out)
        out = self.relu(out)
        out = self.layer3(out)
        return out
```
